```python
import logging
from gensim import corpora, models, similarities
from collections import defaultdict
import pandas as pd  
import string
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

path = "data.csv"
databaru = []

data = pd.read_csv(path, error_bad_lines=False)
databaru = data['reverse_geotag_mediatrac'][1]
documents = databaru

stoplist = set('for a of the and to in'.split())
texts = [[word for word in document.lower().split() if word not in stoplist] 
         for document in documents]
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
texts = [[token for token in text if frequency[token] > 1]
         for text in texts]
from pprint import pprint   # pretty-printer
pprint(texts)
databaru
```

```python
dictionary = corpora.Dictionary(texts)
dictionary.save('/home/robihidayat/Workshop/Python/Project DataBase/listAlamat.dict') # store the dictionary, for future reference
print(dictionary)
dictionary.token2id
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
corpora.MmCorpus.serialize('/home/robihidayat/Workshop/Python/Project DataBase/listAlamat.mm', corpus) # store to disk, for later use
corpus
```

```python
lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
doc = "JL. SROGO - PLANTARAN NO. 5 RT, 03 RW. 04 KALIWUNGU SELATAN KENDAL"
vec_bow = dictionary.doc2bow(doc.lower().split())
vec_lsi = lsi[vec_bow] # convert the query to LSI space
vec_lsi
```

```python
index = similarities.MatrixSimilarity(lsi[corpus]) 
sims = index[vec_lsi]
print(list(enumerate(sims)))
sims = sorted(enumerate(sims), key=lambda item: -item[1])
```
